```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
```

```python
#Download dataframe da GitHub Protezione Civile
df = pd.read_csv('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-province/dpc-covid19-ita-province.csv')

#Selezione dati di Milano e Bergamo
milano = df[df['sigla_provincia'] == 'MI']
bergamo = df[df['sigla_provincia'] == 'BG']
```

------

## Totale casi positivi Milano e Bergamo

```python
# Crea lista di giorni
date = df['data'].unique().tolist()
giorni = []
for n in range(0,len(date)) :
    giorni.append(date[n][5:10])
    
# Crea grafico Bokeh con asse x categorico
casi = figure(plot_width=960,plot_height=640,title='Totale positivi',x_axis_label='giorni',y_axis_label='positivi',x_range=giorni)

casi.line(giorni,milano['totale_casi'],legend_label='Milano',color='blue')
casi.line(giorni,bergamo['totale_casi'],legend_label='Bergamo',color='red')

casi.legend.location = "bottom_right"
output_notebook()

show(casi)
```

------

## Incremento rispetto alla giornata di ieri

```python
#Incremento percentuale rispetto a ieri

ngiorni = 4 #Numero gg da recuperare

#Seleziona il numero di casi degli ultimi due giorni
def casi_ultimi_gg(df) :
    
    i = 1 #Contatore loop
    casi = []
    
    count = df.shape[0]
    col = df.columns.get_loc("totale_casi")
    
    for n in range(0,ngiorni+1):
        casi.append(df.iloc[(count - i),col]) #Casi di oggi
        i = i + 1
    
    return casi

casi_mi = casi_ultimi_gg(milano)
casi_bg = casi_ultimi_gg(bergamo)

def incremento(casi): #parametro casi_mi o casi_bg
    
    inc = []
    for i in range(len(casi)-1):
        rnd = round(((casi[i]/casi[i+1]) * 100) - 100,2)
        if rnd > 0 :
            inc.append('+' + str(rnd) +'%')
        else :
            inc.append(str(rnd) +'%')
    
    return [inc]

inc_mi = incremento(casi_mi)
inc_bg = incremento(casi_bg)

print('Milano:', casi_mi[0]-casi_mi[1], 'nuovi casi rispetto a ieri.\nultimi',ngiorni,'giorni:', inc_mi)
print('')
print('Bergamo:', casi_bg[0]-casi_bg[1], 'nuovi casi rispetto a ieri.\nultimi',ngiorni,'giorni:', inc_bg)
```
